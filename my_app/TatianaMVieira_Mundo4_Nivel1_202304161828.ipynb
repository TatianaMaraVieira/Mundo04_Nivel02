>
> **Estrutura do Projeto**
>
>- **MaterialApp:** Um widget que implementa um aplicativo de material design. Ele fornece várias configurações, como tema, título e roteamento.
>
>- **Scaffold:** Um widget que implementa a estrutura básica de uma tela do aplicativo, fornecendo uma barra de aplicativo, uma barra de navegação e um
corpo que pode conter outros widgets.
>
>- **AppBar:** Uma barra de aplicativo que normalmente contém um título, opções de navegação e ações.
>
>- **TabBar:** Uma barra de guias que exibe várias guias (links) para navegação entre diferentes telas ou seções do aplicativo.
>
>- **TextField:** Um widget de entrada de texto que permite aos usuários inserir texto usando o teclado.
>
>- **BottomAppBar:** Uma barra de aplicativo posicionada na parte inferior da tela, comumente usada para exibir opções de navegação ou ações.
>
>- **Image.asset:** Um widget para exibir imagens locais do projeto. As imagens são carregadas do diretório "images" do projeto.
>
>- **Column:** Um widget que organiza seus filhos em uma coluna vertical, um abaixo do outro.
>
>- **Row:** Um widget que organiza seus filhos em uma linha horizontal, um ao lado do outro.
>
>- **IconButton:** Um botão que exibe um ícone, geralmente usado para ações como pesquisa, compartilhamento, etc.
>
>- **DefaultTabController:** Um controlador de guia que gerencia o estado das guias e a exibição do conteúdo associado a cada guia.
>
>- **TabBarView:** Um widget que exibe o conteúdo associado a cada guia quando a guia correspondente é selecionada.
>
>- **SearchDelegate:** Uma classe que define o comportamento da pesquisa, incluindo ações, sugestões e resultados da pesquisa.
>
