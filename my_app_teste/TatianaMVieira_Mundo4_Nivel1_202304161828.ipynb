># **Missão Prática | Posso criar um App de outra forma!**
>
>
>**Requisitos Funcionais:**
>
>**1. Banner de Destaque:** Um elemento visual destacado, como uma imagem ou
slideshow, exibindo fotos irresistíveis dos destinos oferecidos pela agência. Cada imagem será interativa, direcionando os usuários para páginas específicas de destinos ao serem tocadas.
>
>**2. Barra de Navegação:** Uma barra superior intuitiva, contendo links para diferentes
seções do aplicativo, como "destinos", "pacotes de viagem", "contato" e "sobre nós".Essa navegação facilita o acesso dos usuários às áreas relevantes do app.
>
>**3. Pesquisa Rápida:** Implementação de um campo de pesquisa que permitirá aos
usuários inserir destinos específicos ou dados desejados, agilizando a busca por
pacotes de viagem.  🔎
